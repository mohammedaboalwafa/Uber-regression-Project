```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
from geopy.distance import great_circle
```

```python
df.columns
```

```python
def get_distance(df):
    
    pickup = f'{df["pickup_latitude"]}, {df["pickup_longitude"]}'
    
    destination = f'{df["dropoff_latitude"]}, {df["dropoff_longitude"]}'
    
    try :
        return great_circle(pickup , destination).kilometers
    
    except:
        return np.nan
    
```

```python
df['distance'] = df.apply(get_distance , axis = 1 )
```

```python
df.drop(['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude'] , axis = 1,inplace = True  )
```

```python
from category_encoders import BinaryEncoder
from sklearn.preprocessing import StandardScaler , PolynomialFeatures , MinMaxScaler , RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.metrics import r2_score , mean_squared_error
```

```python
be_col = ['day_name']

be = BinaryEncoder()

df_be = be.fit_transform(df[be_col])

df = pd.concat([df , df_be] , axis = 1 ).drop(be_col , axis = 1 )
```

```python
df.reset_index(inplace = True , drop = True )
```

```python
# x , y = df.drop(['fare_amount'] , axis = 1 ) , df['fare_amount']
x , y = df[['distance']] , df['fare_amount']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2, random_state=42)
```

```python
sclr = RobustScaler()

x_train[x_train.columns] = sclr.fit_transform(x_train)
x_test[x_test.columns] = sclr.transform(x_test)
```

```python
poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(x_train)

x_test_poly = poly.transform(x_test)
```

```python
lr = LinearRegression()

lr.fit(x_train , y_train)

y_train_prd = lr.predict(x_train)
y_test_prd = lr.predict(x_test)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```

```python
sns.scatterplot(x = x_train['distance'] , y = y_train)
sns.lineplot(x = x_train['distance'] , y = y_train_prd , color = 'red')
```

```python
lr = LinearRegression()

lr.fit(x_train_poly , y_train)

y_train_prd = lr.predict(x_train_poly)
y_test_prd = lr.predict(x_test_poly)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```
